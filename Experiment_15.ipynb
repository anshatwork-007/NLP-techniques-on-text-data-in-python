## Title: NLP techniques on text data in python


```python
# tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
text = "Natural language processing (NLP) is a field "
tokens= word_tokenize(text)
print(tokens)
```

```python
#sentence tokenization
from nltk.tokenize import sent_tokenize
text = "python is easy. it is wildly used in data science"
sentences = sent_tokenize(text)
print(sentences)
```

```python
# stop word removal
# stop words such as is, the, and, in are removed
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

words = word_tokenize(text)
filtered_words = [word for word in words if word.casefold() not in stop_words]
print(filtered_words)
```

```python
#stemming
#stemming reduces words to their root form
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ['running', 'runs', 'ran']
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

```python
# lemmetization
# it converts words to their dictionary form
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("cats"))
print(lemmatizer.lemmatize("cacti"))
print(lemmatizer.lemmatize("geese"))
print(lemmatizer.lemmatize("rocks"))
```

```python
# part of speech (POS) tagging
# POS tagging identifies grammatical roles of words

import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag

words = word_tokenize("python is a powerful programming lagnuage")
pos_tags = pos_tag(words)
print(pos_tags)
```

#POS Tag Definitions

01. NNP: Proper noun, singular (e.g., "London", "Microsoft")
02. NN: Noun, singular or mass (e.g., "bicycle", "air")
03. NNS: Noun, plural (e.g., "bicycles", "mountains")
04. VBZ: Verb, 3rd person singular present (e.g., "writes", "takes")
05. VBP: Verb, non-3rd person singular present (e.g., "write", "take")
06. VBD: Verb, past tense (e.g., "wrote", "took")
07. VBG: Verb, gerund or present participle (e.g., "writing", "taking")
08. VBN: Verb, past participle (e.g., "written", "taken")
09. JJ: Adjective (e.g., "green", "large")
10. JJR: Adjective, comparative (e.g., "greener", "larger")
11. JJS: Adjective, superlative (e.g., "greenest", "largest")
12. RB: Adverb (e.g., "quickly", "extremely")
13. DT: Determiner (e.g., "the", "a", "an")
14. PRP: Personal pronoun (e.g., "I", "he", "it")
15. PRP$: Possessive pronoun (e.g., "my", "his", "its")
16. IN: Preposition or subordinating conjunction (e.g., "in", "of", "during")
17. CC: Coordinating conjunction (e.g., "and", "but", "or")
18. CD: Cardinal number (e.g., "one", "2026")
19. MD: Modal (e.g., "can", "should", "will")
